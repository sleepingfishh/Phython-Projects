# Q1

```python
# Import necessary libraries
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

# Initial condition
y0 = 1

# Define dy/dt equation
def dydt(y,t):
    dydt = -2*y*t
    return dydt

# Create the x-axis for the integration
# Time interval is 0 to 2 minutes
start = 0
end = 2
t = np.linspace(start, end, 40)

#Integration of the differential equation
y = sc.integrate.odeint(dydt, y0, t)

# Plotting the graph
plt.figure()
plt.plot(t, y, 'b', label = 'y')
plt.title(f'y(t) vs t')
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True, which = 'both', linestyle = '--', linewidth = 0.1)
plt.minorticks_on()
plt.legend()
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import scipy as sc

# Define coefficient matrix A
A = np.array([[2, 4, 6],[5, 3, 2], [7, 8, 9]])
print(A)

# Define the right-hand side vector b
B = np.array([[18], [13], [30]])
print(B)
```

```python
# Calculate the inverse of A and multiply by b to find the solution vector
C = np.linalg.inv(A).dot(B)
```

```python
# Display the solution vector
print(C)
```
